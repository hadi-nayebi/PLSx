```python
import pandas as pd
import requests
import string
from tqdm import tqdm
```

```python
# 
table_url = lambda letter: f"https://pfam.xfam.org/family/browse?browse={letter}"
```

```python
pages = ["numbers"] + list(string.ascii_lowercase)
print(pages)
```

```python
df = pd.DataFrame()
for c in tqdm(pages):
    r = requests.get(table_url(c)).text
    temp = pd.read_html(r)[1]
    temp.columns = ["ID", "Accession", "Type", "Seed", "Full", "Average length", "Average %id", "Average coverage",
                  "Has 3D", "Change status", "Description"]
    
    df = pd.concat([df, temp])
```

```python
df
```

```python
df.to_csv("../data/pfam/all_famillies.csv")
```
